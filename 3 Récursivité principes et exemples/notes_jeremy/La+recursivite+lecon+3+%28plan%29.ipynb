```python
def facto(n):
    if n <0 :                          #Cas d'erreur
        return "SyntaxError: invalid syntax" 
    if n == 0 :                        # Cas de base
        return 1
    else :
        return n * facto(n-1)        # Appel récursif 
    
    
print(facto(4))

### Pile d'appel
### Début : Phase de descente récursive
|-- facto(4)
|  |-- facto(3)
|  |  |-- facto(2)
|  |  |  |-- facto(1)
|  |  |  |  |-- facto(0) ### Fin : Phase de descente récursive
|  |  |  |  |  |-- return 1 ### Début : Phase de remontée
|  |  |  |  |-- return 1
|  |  |  |-- return 2
|  |  |-- return 6
|  |-- return 24 ### Fin :Phase de remontée
24   # Valeur retourné
###
```

```python
"Trois fonctions pour calculer u_n"
def u1(n):
    "Version itérative"
    u = 2
    for i in range(n):
        u = 0.5 * (u + 1 / u)
    return u 

On fait trois opérations dans chaque itération donc C_1(n)=3n . 
La complexité est linéaire (très bonne)

def u2(n):
    "Version récursive basique"
    if n<0 :
        return "n need to be positif ;)"
    if n == 0:
        return 2
    return 0.5 * (u2(n - 1) + 1 / u2(n - 1)) 
On a C2(0)=0 et  pour n∈N∗,  C_2(n)=2C2(n−1)+3. On en déduit (récurrence facile) que C_ 2(n)=3(2^n−1)~=O(2^n) .
La complexité est géométrique (rédhibitoire). 

def u3(n):
    "Version récursive "
    if n<0 :
        return "n need to be positif ;)"
    if n == 0:
        return 2
    x = u3(n-1)
    return 0.5 * (x + 1 / x) 

On a C3(0)=0 et  et  pour  n∈N∗, C3(n)=C3(n−1)+3. On en déduit que C3(n)=3n. 
La complexité est linéaire (très bonne).

#####Bilan:
    - Les versions u1(n) et u3(n) ont la même complexité. La version u3(n) consomme plus de mémoire que u1(n), mais cela reste négligeable.
    - La version u2(n), la plus naturelle, est à proscrire absolument. 
```

```python
# -Suite de Fibonacci
Etudions maintenant la complexité des trois versions suivantes du calcul de la suite de Fibonacci. 
Notons C(n) le nombre d’opérations arithmétiques élémentaires nécessaires au calcul de Fn.

def fibo1(n):
    a, b = 1, 1
    for i in range(2, n + 1):
        a, b = b, a + b
    return b 

On fait une addition par itération, donc C1(n)=n−1.  
La complexité est linéaire (très bonne).

def fibo2(n):
    if n < 2:
        return 1
    return fibo2(n-1) + fibo2(n-2) 
On a C2(0)=C2(1)=0 et  pour n ≥ 2,  C2(n)=C2(n−1)+C2(n−2)+1. 

On en déduit (calcul classique) que C2(n)=1/√5 * (ϕ^(n+1)+1/ϕ^(n+1)−1 (avec ϕ=(1+√5)/2≈1.6) .
La complexité est géométrique (rédhibitoire).
                                                  
def fibo3(n):
    "On propose alors la version suivante, la fonction fibo3(n) enveloppe le moteur de calcul récursif"
    def moteur(k, a,b):
        if k == 0:
            return a
        return moteur(k - 1, b, a + b)
    return moteur(n, 1, 1)                                              
 
####Bilan:
    - La version fibo2(n), la plus naturelle, est à proscrire absolument.
    - Les versions fibo1(n) et fibo3(n) ont la même complexité.
    - Pour écrire une version récursive efficace fibo3(n), 
il faut travailler avec deux termes consécutifs de la suite.  
```
